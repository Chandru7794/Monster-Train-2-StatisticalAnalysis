## Section 1 — W/L by Clan

Three views of clan win rate: primary role only, secondary role only, and either role combined. The combined view is for reference only — because runs appear in both primary and secondary counts, it cannot be used for statistical testing.

Clans are ordered by the canonical project order (Banished → Railforged), not by win rate.

### 1a — W/L by Primary Clan

Win rate for each clan when used as the primary (driving) clan. This is the most meaningful role split — the primary clan defines the core deck strategy.

### 1b — W/L by Secondary Clan

Win rate for each clan when used as the secondary (support) clan. Compared against 1a, this reveals whether clans perform differently depending on their role — a key pattern already observed in the SQL EDA (e.g. Banished strong primary/weak secondary, Pyreborne the reverse).

### 1c — W/L by Clan, Either Role

Combined win rate regardless of role. Useful for seeing total appearances and overall record, but not suitable for statistical testing since each run contributes to two clans.

## Section 2 — Confounded Secondary Performance

Two clans showed surprisingly poor secondary win rates in the SQL EDA: Banished and Underlegion. Both were frequently paired with Pyreborne as primary. These charts test whether the poor secondary performance is specific to Pyreborne pairings or consistent across all primaries.

Each chart collapses to two bars: runs where Pyreborne was primary vs. all other primaries combined.

### 2a — Banished Secondary: Pyreborne Primary vs. All Other Primaries

Banished as secondary is notably weak overall. Given that Banished + Pyreborne was the most-played combination (15.8% of all runs, driven by early availability), this chart checks whether Pyreborne primary is the confounding factor.

### 2b — Underlegion Secondary: Pyreborne Primary vs. All Other Primaries

Underlegion is strong as primary but surprisingly poor as secondary. Underlegion + Pyreborne is the second most-played combination (7.9% of runs). This chart tests whether that pairing is driving the weak secondary record.

## Section 3 — Learning Curve: Rolling Win Rate Over Time

Win rate smoothed over a 10-run rolling window, plotted by run number. The SQL segment analysis showed clear improvement from Dec 2025 → Jan 2026 → Feb 2026. This chart shows the same trend continuously rather than in fixed buckets, making it easier to see when the game started clicking and whether improvement has continued.

## Section 6 — Covenant Level Distribution Over Time (Rug Plot)

A rug plot showing when each covenant level was played across the run history. Each tick mark represents one run, grouped by covenant level on the y-axis and plotted by run date or run number on the x-axis.

This reveals how covenant level progressed over time — when harder difficulties were introduced, and whether certain covenant levels cluster in specific time periods.
